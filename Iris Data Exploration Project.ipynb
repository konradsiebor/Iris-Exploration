```
iris = pd.read_csv('IRIS.csv')
```

```
iris.head()
```

**Question 1: How many of each flower are there?**

```
iris['species'].value_counts()
```

**The data set is evenly split between the three different species of iris (50 of each)**

**Question 2: What is the average sepal area?**

```
round(iris['sepal_length'].mean()*iris['sepal_width'].mean(), 2)
```

**The average pedal length is 17.85 cm^2**

**Question 3: What is the average petal area?**

```
round(iris['petal_length'].mean()*iris['petal_width'].mean(), 2)
```

**The average pedal length is 4.51 cm^2**

**Question 3: Which species has the largest sepal length?**

```
iris['sepal_length'].groupby(iris['species']).mean()
```

**Looks like virginica, following by versicolor, and finally setosa**

****Question 4: Which species has the smallest petal to sepal length ratio?****

```
(iris['petal_length'].groupby(iris['species']).mean())/(iris['sepal_length'].groupby(iris['species']).mean())
```

**Setosa by far**
